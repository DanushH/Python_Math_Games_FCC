
# Scatter plot game

```python
import matplotlib.pyplot as plt
import random

def play_scatter_plot_game():
    score = 0
    xmin = -20
    xmax = 20
    ymin = -20
    ymax = 20

    fig, ax = plt.subplots()

    for i in range(0,3):
        xpoint = random.randint(xmin, xmax)
        ypoint = random.randint(ymin, ymax)
        x = [xpoint]
        y = [ypoint]
        plt.axis([xmin,xmax,ymin,ymax])
        plt.plot([xmin,xmax],[0,0],'g')
        plt.plot([0,0],[ymin,ymax], 'g')
        plt.plot(x, y, 'ro')
        print(" ")
        plt.grid()
        plt.show()
        guess = input("What are the coordinates of the red point? Ex: 0,0 \n")
        guess_array = guess.split(",")
        xguess = int(guess_array[0])
        yguess = int(guess_array[1])
        if xguess == xpoint and yguess == ypoint:
            print(f"Correct, x = {xpoint} y = {ypoint}")
            score = score + 1
        else:
            print(f"Incorrect, x = {xpoint} y = {ypoint}")

    return score


score = play_scatter_plot_game()
print(f"\nYour total score: {score}")
```

# **Algebra practice game**

```python
import random

def one_step_problem():
    x = random.randint(-100, 100)
    y = random.randint(-100, 100)
    operator = random.choice(['+', '-', '*', '/'])

    question = f"{x} {operator} {y}"
    answer = round(eval(question), 2)

    return f"x {operator} {y} = {answer}", x


def two_step_problem():
    x = random.randint(-100, 100)
    y = random.randint(-100, 100)
    z = random.randint(-100, 100)
    operator1 = random.choice(['+', '-', '*', '/'])
    operator2 = random.choice(['+', '-', '*', '/'])

    question = f"{x} {operator1} {y} {operator2} {z}"
    answer = round(eval(question), 2)

    return f"x {operator1} {y} {operator2} {z} = {answer}", x


def main():
    score = 0
    question_count = 5

    print("Algebra Practice Game\n")

    for _ in range(question_count):
        question_type = random.choice([1, 2])

        if question_type == 1:
            question, x = one_step_problem()
        else:
            question, x = two_step_problem()

        print(question)
        user_answer = int(input("Find x:  "))
        if user_answer == x:
            print(f"Great, x = {x}\n")
            score += 1
        else:
            print(f"Nice effort, x is actually {x}\n")

    print(f"Your total score: {score}")


if __name__ == "__main__":
    main()
```
